```python
import PyPDF2 as pdf
```

```python
pdf_reader = pdf.PdfReader(file)
```

```python
len(pdf_reader.pages)
```

```python
page_1 = pdf_reader.pages[0]
```

```python
print(page_1)
```

```python
page_1.extract_text()
```

```python
pdf_writer = pdf.PdfWriter()
```

```python
output = open("new_sample.pdf","wb")
```

```python
pdf_writer.add_page(page_1)
```

```python
pdf_writer.write(output)
```
