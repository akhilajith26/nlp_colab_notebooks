```python
 import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from numpy.random import randint
```

```python
x = np.linspace(0,10,20)
```

```python
x
```

```python
y = randint(0,50,20)
```

```python
y
```

```python
y = np.sort(y)
```

```python
y
```

```python
plt.plot(x,y,color = 'm',linestyle = '--',marker = '*',markersize = 10,linewidth = 2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y axis')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='r')
plt.subplot(1,2,2)
plt.plot(x,y,color='g')
```

```python
fig= plt.figure() #create a fig object
axes= fig.add_axes([0.1,0.1,1,1]) #add axes to fig object which will create a blank canvas
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('X vs Y axis')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(x,y,color='r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Plot 1')

ax2.plot(x,y,color='g')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Plot 2')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
y_views=[534,690,258,402,724,689,353]
f_views=[123,342,700,305,406,648,325]
t_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]
```

```python
plt.scatter(days,y_views,label = 'Youtube Views',marker='o')
plt.scatter(days,f_views,label = 'Facebook Views',marker='o')
plt.scatter(days,t_views,label = 'Twitter Views',marker='o')
plt.xlabel("Days")
plt.ylabel("Views")
plt.title("Socal media views")
plt.grid(color='r',linestyle='-')
plt.legend(loc=1)
```

```python
plt.bar(days,y_views,label="Youtube Views")
plt.bar(days,f_views,label="Facebook Views")
plt.xlabel("Days")
plt.ylabel("Views")
plt.title("Socal media views")
plt.legend()
```
